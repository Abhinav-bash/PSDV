### 1. List Operations and Basic Statistics: 

```python
NaturalNo = []
i = 0
while i < 20:
    i += 1
    NaturalNo.append(i)
sumEven = 0
for k in NaturalNo :
    if k%2 == 0:
        sumEven = sumEven + k
productOdd = 1
for j in NaturalNo :
    if j%2 == 1 :
        productOdd = productOdd*j
mean = sum(NaturalNo)/20
median = (NaturalNo[9] + NaturalNo[10])/2
# Below code is for standard deviation
DeviationSquare = 0
SumSquareDeviation = 0
for m in NaturalNo :
    DeviationSquare = (m-mean)**2
    SumSquareDeviation += DeviationSquare
standardDeviation = (SumSquareDeviation/20)**0.5

print(f"sum of all even numbers is {sumEven}")
print(f"product of all odd numbers is {productOdd}")
print(f'mean is {mean}, median is {median}, and standard deviation is {standardDeviation}')
```

### 3. Fibonacci Sequence: 

```python
import matplotlib.pyplot as graph
n = int(input())
def Fibonacci(n):
    
    FibbonacciSequence = [0,1]
    counter = 0
    firstTerm = 0
    secondTerm = 1
    while counter < n-2:
        TempTerm = firstTerm + secondTerm
        firstTerm = secondTerm
        secondTerm = TempTerm
        FibbonacciSequence.append(TempTerm)
        counter += 1
    return FibbonacciSequence

# Below we will deal with plotting
x_axis = []
for i in range (1,n+1):
    x_axis.append(i)
y_axis = Fibonacci(n)
graph.plot(x_axis,y_axis, color = 'pink')
graph.xlabel('Term')
graph.ylabel('Fibonacci Sequence')
graph.title('Fibonacci Sequence of n numbers')
graph.show()
```

### 7. Dictionary Operations:

```python
import matplotlib.pyplot as graph
StudentMarks = {
    "Walter" : 90,
    "jessie" : 80,
    "Mike" : 60,
    "Gus" : 100
}
marks = list(StudentMarks.values())


avgMarks = sum(marks)/len(marks)
print(f'Average Marks of Students is {avgMarks}')

StudentsAboveAvg = []
for student, mark in StudentMarks.items() :
    if mark > avgMarks :
        StudentsAboveAvg.append(student)
above_avg_students = [student for student, mark in StudentMarks.items() if mark > avgMarks]

print("Students who scored above average:", ", ".join(above_avg_students))
Students = list(StudentMarks.keys())
Marks = list(StudentMarks.values())

graph.bar(Students, Marks, color = 'pink')
graph.show()
```

### 9. Merging Dictionaries:

```python
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 15, 'c': 25, 'd': 35}
merged_dict = dict1.copy()
for key, value in dict2.items():
    if key in merged_dict:
        merged_dict[key] += value 
    else:
        merged_dict[key] = value
print(merged_dict)
  
```

### 10. Character Frequency:

### 12. String Operations:

```python
import matplotlib.pyplot as graph
GivenList = ['apple','banana','chappal','science','apple','pear','computer','laptop','science']

# Below code is for longest string
longestString = 0
for element in GivenList:
    if len(element) > longestString :
        longestString = len(element)
        LongestIndex = GivenList.index(element)
print(f'Longest string is of length {longestString} and String is {GivenList[LongestIndex]}')

# Below code is to find no. of strings that start with a vowel
vowelsCount = 0
vowels = ['a','e','i','o','u']
for words in GivenList:
    if words[0] in vowels:
        vowelsCount+=1
print(f'Strings that start with vowels are {vowelsCount}')

# Below code is to visualize lenght of String on an histogram
x_axis = [i for i in range(1,len(GivenList))]
y_axis = []
for i in GivenList:
    y_axis.append(len(i))
graph.title('Words and Count')
graph.xlabel('Count')
graph.ylabel('Words')
graph.hist(y_axis, color='pink' )
```

### 13. Sorting Dictionaries:

### 22. Normalization: 

```python
import matplotlib.pyplot as graph
NumberList = [1,3,4,5,7,9,2,3,6,11,14,16]
numbers = len(NumberList)
maxNumber = max(NumberList)
minNumber = min(NumberList)
print(maxNumber)
elementsInList = []
for i in range (1, numbers +1) :
    elementsInList.append(i)
NormalizedList = []
for j in NumberList :
    x = (j - minNumber)/(maxNumber - minNumber)
    NormalizedList.append(x)
x_axis = elementsInList
graph.plot(x_axis, NumberList, label = ' Original List ', marker='o', linestyle='-', color = 'red')
graph.plot(x_axis, NormalizedList, label = ' Normalized List ',marker='o', linestyle='-', color = 'pink')
graph.xlabel('Term')
graph.ylabel('Normalized Number')
graph.title('Normalization')
graph.legend()
graph.grid(True)
graph.xticks(x_axis)
graph.yticks(NumberList)
graph.yticks()
graph.show()
```

### 25. Leap Year Checker: 

```python
year = int(input())
# Every year that is exactly divisible by four is a leap year, except for years that are exactly divisible by 100,
# but these centurial years are leap years if they are exactly divisible by 400
if  year % 400 == 0 :
    print(f'{year} is leap year')
elif year % 100 == 0 :
    print(f'{year} is not leap year')
elif year % 4 == 0 :
    print(f'{year} is leap year')
else :
    print(f'{year} is not leap year')
```
